```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sorted =df.sort_values(by="Mar Cap - Crore",ascending=False).head(10)


# Extracting the top 10 companies
top_10_companies = df_sorted

# Creating the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_companies['Name'], top_10_companies['Mar Cap - Crore'], color='skyblue')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Market Capitalization')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest market cap at the top
  

# Add data labels

plt.tight_layout()
plt.show()
```

```python
#data sorted by quarterly sales

df.sort_values(by="Sales Qtr - Crore",ascending=False).head(10)
```

Quarterly sales is more in I O C L with 110666.93 Crore and then comes Reliance Inds. with 99810.00 Crores

```python
df_sorted_2 = df.sort_values(by="Sales Qtr - Crore", ascending=False).head(10)

# Extracting the top 10 companies
top_sales = df_sorted_2

# Creating the horizontal bar chart
plt.figure(figsize=(15, 6))
sns.barplot(x=top_sales['Name'], y=top_sales['Sales Qtr - Crore'], color='skyblue')
plt.xlabel('Company Name')
plt.ylabel('Quarterly Sales')
plt.title('Top 10 Companies by Quarterly Sales')

# Invert y-axis to have the highest market cap at the top
  
# Add data labels
plt.show()
```

```python


# to see top 10 companies 

df.sort_values(by=["Mar Cap - Crore","Sales Qtr - Crore"],ascending=False).head(10)
```

```python
plt.figure(figsize=(15, 8))  

# Sorting the DataFrame
df_sorted = df.sort_values(by="Mar Cap - Crore", ascending=False).head(10)
df_sorted_2 = df.sort_values(by="Sales Qtr - Crore", ascending=False).head(10)

# Extracting top companies
top_sales = df_sorted_2
top_10_companies = df_sorted

# Plotting market capitalization on the primary y-axis
plt.bar(top_10_companies["Name"], top_10_companies["Mar Cap - Crore"], color='b')
plt.ylabel('Market Capitalization (Crore)', color='b')
plt.tick_params(axis='y', colors='b')

# Creating a secondary y-axis for quarterly sales
plt.twinx()
plt.plot(top_10_companies["Name"], top_10_companies["Sales Qtr - Crore"], color='r', marker='o')
plt.ylabel('Quarterly Sales (Crore)', color='r')
plt.tick_params(axis='y', colors='r')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, fontsize=10)

# Title and labels
plt.title('Market Capitalization and Quarterly Sales')
plt.xlabel('Company')

plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(data=df.corr())
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df["Sales Qtr - Crore"],df["Mar Cap - Crore"] , alpha=0.5)

# Annotate each point with the corresponding company name

plt.title('Relationship between Market Capitalization and Quarterly Sales')
plt.xlabel('Quarterly Sales (Crore)')
plt.ylabel('Market Capitalization (Crore)')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Histogram for Market Capitalization
sns.histplot(df['Mar Cap - Crore'], ax=axes[0], kde=True, color='skyblue')
axes[0].set_title('Distribution of Market Capitalization')
axes[0].set_xlabel('Market Capitalization(In Crores)')
axes[0].set_ylabel('Frequency')

# Histogram for Quarterly Sales
sns.histplot(df['Sales Qtr - Crore'], ax=axes[1], kde=True, color='salmon')
axes[1].set_title('Distribution of Quarterly Sales')
axes[1].set_xlabel('Quarterly Sales(In Crores)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Most number of companies have market capitalization below 100000 cores and quarterly sales below 20000 crores.
